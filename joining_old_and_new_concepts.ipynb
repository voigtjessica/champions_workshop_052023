# Joining the old concepts with new concepts

```python
import pandas as pd

# Abrindo o arquivos
df_novo = pd.read_csv("C:\\Users\\JVoigt\\OneDrive - Universität für Weiterbildung Krems\\Dokumente\\Python Scripts\\Champions_Workshop\\concepts_last.csv",sep = ';')
df_velho = pd.read_excel("C:\\Users\\JVoigt\\OneDrive - Universität für Weiterbildung Krems\\Dokumente\\Python Scripts\\Champions_Workshop\\concepts_normalization.xlsx",
                         sheet_name = 'coding')

# print('df velho: \n', df_velho.head(1), '\n df novo: \n', df_novo.head(1))
# Key: statement_id 
```

```python
velhos = df_velho.statement_id.to_list()
novos = df_novo.statement_id.to_list()
novos_filtrados = [x for x in novos if x not in velhos]  # df

#len(novos_filtrados)   # parece que deu certo
df_filtrado = df_novo[df_novo['statement_id'].isin(novos_filtrados)]

df_final = pd.concat([df_velho,df_filtrado],ignore_index=True)
```

```python
# Second part: cleaning the data

df_final_clean = df_final

#1. Removing empty concepts:
df_final_clean = df_final_clean[df_final_clean['concept'].notnull()]   #2 removed

#2. Verify where person and or organization is empty and manually fix it:

# df_temp = df_final_clean[df_final_clean['person'].isnull() | df_final_clean['organization'].isnull()]
# df_temp

# 177: Georg Strasser övp
# 362 :Josef Hechenberger övp

df_final_clean.loc[df_final_clean['statement_id'] == 177, 'person'] = 'Georg Strasser'
df_final_clean.loc[df_final_clean['statement_id'] == 177, 'organization'] = 'övp'

df_final_clean.loc[df_final_clean['statement_id'] == 362, 'person'] = 'Josef Hechenberger'
df_final_clean.loc[df_final_clean['statement_id'] == 362, 'organization'] = 'övp'

#df_final_clean.query('statement_id == 177') # OK

#3 . Verify if Agreement is empty:
# df_temp = df_final_clean[df_final_clean['agreement'].isnull()] # Empty
```
